```python
dic_dp = {0:0,1:1}
def fib_dp(n):
  if n in dic_dp:
    return dic_dp[n]
  else:
    dic_dp[n] = fib_dp(n-1) + fib_dp(n-2)
    return dic_dp[n]
```

```python
start_dp = timeit.default_timer()
print(fib_dp(20))
stop_dp = timeit.default_timer()
print('Time DP: ', stop_dp - start_dp) 
```
