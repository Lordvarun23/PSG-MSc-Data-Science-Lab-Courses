```python
def f(x):
    return x*x*(25-x*x)

a = 1
h = 1
k = a * 0.5 * h * h

t_max = 3
x_max = 5

col = int(x_max / h) + 1
row = int(t_max / k) + 1

ans = np.zeros((row,col))

for i in range(row):
    for j in range(col):
        if i == 0 and j != 0 and j != (col-1):
            ans[i][j] = f(j)
            continue

        if i != 0 and j != 0 and j != (col-1):
            a = ans[i-1][j-1]
            b = ans[i-1][j+1]

            ans[i][j] = (a+b) * 0.5

print(ans)
```
